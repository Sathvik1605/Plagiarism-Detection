#  Tokenizing and removing stopwords

```python
token_ds=text_ds.copy()
from nltk.tokenize import word_tokenize
c=-1
for i in text_ds["Text"]:
    c=c+1
    token_ds["Text"][c]=word_tokenize(i)
print(token_ds["Text"][0])
```

```python
#removing stop words
from nltk.corpus import stopwords
a=set(stopwords.words('english')).union(",","(",")","[","]","{","}","#","@","!",":",";",".","?")
punc='''#!@$%^&*()_+-={}[]|\:;"',.<>?/`~/*'''
c=-1
#removing stop words
for i in range(len(token_ds)):
    for j in token_ds["Text"][i]:
        if j.lower() in a:
            token_ds["Text"][i].remove(j)
#removing punctuations
for i in range(len(token_ds)):
    for j in token_ds["Text"][i]:
        if j in punc:
            token_ds["Text"][i].remove(j)
```

```python
token_ds
```

# Casefolding 

```python
cf_ds=token_ds.copy()
for i in range(len(cf_ds)):
    cf=[]
    for w in cf_ds["Text"][i]:
        cf.append(w.casefold())
    cf_ds["Text"][i].clear()
    cf_ds["Text"][i]=cf.copy()
```

```python
cf_ds
```

# Lemmatization

```python
lem_ds=cf_ds.copy()
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for i in range(len(lem_ds)):
    lem=[]
    for w in lem_ds["Text"][i]:      
        lem.append(lemmatizer.lemmatize(w))
        print(w, " : ", lemmatizer.lemmatize(w))
    lem_ds["Text"][i].clear()
    lem_ds["Text"][i]=lem.copy()
```

```python
lem_ds
```

#  Inverted indexing

```python
dict={}
for i in range(len(lem_ds)):
    s=[]
    for w in lem_ds["Text"][i]:
        if w not in s:
            s.append(w)
            if w not in dict.keys():
                dict[w]=[]
            dict[w].append(i)
    
```

# Calculating cosine similarity of Dataset w.r.t original documents

```python
cs_ds=lem_ds.copy()
```

```python
cs_ds["coscine_similarity"]=""
cs_ds
```

```python
import math
c=0
result=[]
for i in range(len(cs_ds)):
    k=i%5
    l=95+k
    sum=0
    deno_1=0
    deno_2=0
    for j in all_words:
        if(i in dict[j]):
            word_freq_1=cs_ds["Text"][i].count(j)
            deno_1=deno_1+word_freq_1*word_freq_1
        if(l in dict[j]):
            word_freq_2=cs_ds["Text"][l].count(j)
            deno_2=deno_2+word_freq_2*word_freq_2
        if(i in dict[j] and l in dict[j]):
            sum=sum+word_freq_1*word_freq_2
    if(sum!=0):
        result.append(sum/(math.sqrt(deno_1)*math.sqrt(deno_2)))
    else:
        result.append(0)
           

print(result)
result.count(0)
```

# Plagiarism detection based on tf-idf value

```python
tf_idf_ds=lem_ds.copy()
tf_idf_ds["tf-idf"]=""
idf={}
for i in dict: #calculating idf of each word and storing in a dictionary
    idf[i]=((math.log(100/(len(dict[i])+1),10)))
```

```python
idf
```

```python
dict_tf_idf={}
for i in range(len(tf_idf_ds)):
    dict_tf_idf[i]=[]
    for j in all_words:
        w=[]
        if(i in dict[j]):
            word_freq=tf_idf_ds["Text"][i].count(j)
            tf=math.log((word_freq/len(tf_idf_ds["Text"][i]))+1,10)#tf of each word in document i
            #w[j]=(tf*idf[j]) #tf-idf calculation for all words in a document
            w=[j,tf*idf[j]] #tf-idf calculation for all words in a document
            dict_tf_idf[i].append(w)#tf-idf value for all words in each document
```
